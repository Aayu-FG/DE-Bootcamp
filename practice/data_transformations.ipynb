```python
import pandas as pd
```

```python
sales = [100,130,119,92,35]
customer_account = ['B100','J101','X102','P103','R104']
city = ['BOS','LA','NYC','SF','CHI']
```

```python
#Create a DataFrame with the data above:
df = pd.DataFrame({
    'sales': sales,
    'customer_account': customer_account,
    'city': city
})
```

```python
print(df)
```

```python
#What is the name of the first column?:
df.columns[0]
```

```python
#Sort the DataFrame by city in descending order:
df_sorted = df.sort_values(by='city', ascending=False)
print(df_sorted)
```

```python
#Which customer is in the last row of the DataFrame?:
df.iloc[-1]['customer_account']
```

```python
#For sorted:
df_sorted.iloc[-1]['customer_account']
```

```python
#Reorder the columns with customer first:
df_reordered = df[['customer_account', 'sales', 'city']]
print(df_reordered)
```
