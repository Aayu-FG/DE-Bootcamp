```python
#Error handling : Check for file

filename = "products.csv"
try:
    if not os.path.exists(filename):
        raise FileNotFoundError(f"File {'filename'} does not exist")
    with open(filename, 'r') as file:
        content = file.read()
    
except FileNotFoundError as e:
    print(f"Error: {e}")
except PermissionError:
    print(f"Error: Permission denied to read '{filename}'")
except Exception as e:
    print(f"Error: Unexpected error {e} ")
```

```python
#Example 1

try:
    with open('products.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)
        
        for row in reader:
            if len(row) < 8:  
                print(f"Skipping row with missing data: {row}")
                continue

            product_id, product_name, category, price, stock_quantity, supplier, rating, description = row
            print(f"{product_name} | ₹{price} | Stock: {stock_quantity} | Supplier: {supplier}")

except FileNotFoundError:
    print("Error: The 'products.csv' file could not be found.")
except csv.Error as e:
    print(f"CSV Error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")

```
